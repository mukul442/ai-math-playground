## Ques 7-16 - Simplify the rational expression

```python
exprs = [
    cancel((3*(x + 2)*(x - 1)) / (6*(x - 1)**2)),
    cancel((4*(x**2 - 1)) / (12*(x + 2)*(x - 1))),
    cancel((x - 2) / (x**2 - 4)),
    cancel((x**2 - x - 2) / (x**2 - 1)),
    cancel((x**2 + 6*x + 8) / (x**2 + 5*x + 4)),
    cancel((x**2 - x - 12) / (x**2 + 5*x + 6)),
    cancel((y**2 + y) / (y**2 - 1)),
    cancel((y**2 - 3*y - 18) / (2*y**2 + 5*y + 3)),
    cancel((2*x**3 - x**2 - 6*x) / (2*x**2 - 7*x + 6)),
    cancel((1 - x**2) / (x**3 - 1)),
]

def gen_exprs(exprs, ques_start):
    for i, expr in enumerate(exprs, start=ques_start):
        display(Math(f"{i}: {latex(expr)}"))

gen_exprs(exprs, ques_start=7)
```
