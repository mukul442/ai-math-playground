### Ques. 109. Degrees of Sums and Products of Polynomials

```python
expr_pairs = [
    (x**2 + 2*x + 16, x**2 + 4*x + 20),
    (2*x**3 + x**2 + 2*x + 16, x**2 + 4*x + 20),
    (2*x**3 + x**2 + 2*x + 16, x**3 + x**2 + 4*x + 20),
]

# a. For product of two polynomials the degree of product is sum of degrees of polynomials.
for i, expr in enumerate(expr_pairs):
    display(Math(f"{i}: {latex(expand(expr_pairs[i][0] * expr_pairs[i][1]))}"))

# b. The sum of two polynomials is the same degree as of original one.
for i, expr in enumerate(expr_pairs):
    display(Math(f"{i}: {latex(expand(expr_pairs[i][0] + expr_pairs[i][1]))}"))
```
